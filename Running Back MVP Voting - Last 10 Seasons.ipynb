<h2>Import libraries

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

<h2>Import data & create dataframes for each season; also drop 1st level of multi-level index, add a "Season" column for the appropriate season, and subset each dataframe to the necessary columns; scoring changed in 2022, so had to rename "1st Place" to "Votes" to match other dataframes.

```python
url_2013 = "https://www.pro-football-reference.com/awards/awards_2013.htm"
mvp_2013 = pd.read_html(url_2013)
mvp_2013 = mvp_2013[0]
mvp_2013.columns = mvp_2013.columns.droplevel(0)
mvp_2013["Season"] = 2013
mvp_2013 = mvp_2013[["Season", "Rk", "Pos", "Player", "Tm", "Votes", "Yds", "TD"]]

url_2014 = "https://www.pro-football-reference.com/awards/awards_2014.htm"
mvp_2014 = pd.read_html(url_2014)
mvp_2014 = mvp_2014[0]
mvp_2014.columns = mvp_2014.columns.droplevel(0)
mvp_2014["Season"] = 2014
mvp_2014 = mvp_2014[["Season", "Rk", "Pos", "Player", "Tm", "Votes", "Yds", "TD"]]

url_2015 = "https://www.pro-football-reference.com/awards/awards_2015.htm"
mvp_2015 = pd.read_html(url_2015)
mvp_2015 = mvp_2015[0]
mvp_2015.columns = mvp_2015.columns.droplevel(0)
mvp_2015["Season"] = 2015
mvp_2015 = mvp_2015[["Season", "Rk", "Pos", "Player", "Tm", "Votes", "Yds", "TD"]]

url_2016 = "https://www.pro-football-reference.com/awards/awards_2016.htm"
mvp_2016 = pd.read_html(url_2016)
mvp_2016 = mvp_2016[0]
mvp_2016.columns = mvp_2016.columns.droplevel(0)
mvp_2016["Season"] = 2016
mvp_2016 = mvp_2016[["Season", "Rk", "Pos", "Player", "Tm", "Votes", "Yds", "TD"]]

url_2017 = "https://www.pro-football-reference.com/awards/awards_2017.htm"
mvp_2017 = pd.read_html(url_2017)
mvp_2017 = mvp_2017[0]
mvp_2017.columns = mvp_2017.columns.droplevel(0)
mvp_2017["Season"] = 2017
mvp_2017 = mvp_2017[["Season", "Rk", "Pos", "Player", "Tm", "Votes", "Yds", "TD"]]

url_2018 = "https://www.pro-football-reference.com/awards/awards_2018.htm"
mvp_2018 = pd.read_html(url_2018)
mvp_2018 = mvp_2018[0]
mvp_2018.columns = mvp_2018.columns.droplevel(0)
mvp_2018["Season"] = 2018
mvp_2018 = mvp_2018[["Season", "Rk", "Pos", "Player", "Tm", "Votes", "Yds", "TD"]]

url_2019 = "https://www.pro-football-reference.com/awards/awards_2019.htm"
mvp_2019 = pd.read_html(url_2019)
mvp_2019 = mvp_2019[0]
mvp_2019.columns = mvp_2019.columns.droplevel(0)
mvp_2019["Season"] = 2019
mvp_2019 = mvp_2019[["Season", "Rk", "Pos", "Player", "Tm", "Votes", "Yds", "TD"]]

url_2020 = "https://www.pro-football-reference.com/awards/awards_2020.htm"
mvp_2020 = pd.read_html(url_2020)
mvp_2020 = mvp_2020[0]
mvp_2020.columns = mvp_2020.columns.droplevel(0)
mvp_2020["Season"] = 2020
mvp_2020 = mvp_2020[["Season", "Rk", "Pos", "Player", "Tm", "Votes", "Yds", "TD"]]

url_2021 = "https://www.pro-football-reference.com/awards/awards_2021.htm"
mvp_2021 = pd.read_html(url_2021)
mvp_2021 = mvp_2021[0]
mvp_2021.columns = mvp_2021.columns.droplevel(0)
mvp_2021["Season"] = 2021
mvp_2021 = mvp_2021[["Season", "Rk", "Pos", "Player", "Tm", "Votes", "Yds", "TD"]]

url_2022 = "https://www.pro-football-reference.com/awards/awards_2022.htm"
mvp_2022 = pd.read_html(url_2022)
mvp_2022 = mvp_2022[0]
mvp_2022.columns = mvp_2022.columns.droplevel(0)
mvp_2022["Season"] = 2022
mvp_2022 = mvp_2022[["Season", "Rk", "Pos", "Player", "Tm", "1st Place", "Yds", "TD"]]
mvp_2022 = mvp_2022.rename(columns={"1st Place": "Votes"})
```

<h2>Concatenate all dataframes to get 1 big dataframe.

```python
mvp_df = pd.concat([mvp_2013,mvp_2014,mvp_2015,mvp_2016,mvp_2017,mvp_2018,mvp_2019,mvp_2020,mvp_2021,mvp_2022])
display(mvp_df)
```

<h2>Rename "Rk" to "Rank" and fill NaN values with 0.

```python
mvp_df = mvp_df.rename(columns={"Rk": "Rank"})
mvp_df = mvp_df.fillna(0)
display(mvp_df)
```

<h2>Query dataframe for running backs only

```python
mvp_df = mvp_df.query('Pos == "RB" ')
display(mvp_df.sort_values("Rank"))
```

<h2>Use index to rename duplicate "Yds" and "TD" columns to the appropriate category

```python
mvp_df.columns = ["Season","Rank","Pos","Player","Tm","Votes","Pass_Yds","Rush_Yds","Rec_Yds","Pass_TDs","Rush_TDs","Rec_TDs"]
display(mvp_df.sort_values("Rank"))
```

<h2>Subset dataframe for necessary columns, create "Total_Yds" and "Total_TDs" columns

```python
mvp_df = mvp_df[["Season", "Rank", "Pos", "Player", "Tm", "Votes", "Rush_Yds", "Rec_Yds", "Rush_TDs", "Rec_TDs"]]
mvp_df["Total_Yds"] = mvp_df["Rush_Yds"] + mvp_df["Rec_Yds"]
mvp_df["Total_TDs"] = mvp_df["Rush_TDs"] + mvp_df["Rec_TDs"]
display(mvp_df.sort_values("Rank"))
```

<h2>Limit only to rows that rank in top 5

```python
mvp_df = mvp_df[mvp_df.Rank <= 5]
display(mvp_df.sort_values("Rank"))
```
